```python
#Importing the libraries i'm going to use
import pandas as pd
import numpy as np
import matplotlib.pyplot  as plt
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
```

# Ενδιάμεση Εργασία 

First of all i am going  to load the data into a pandas dataframe and visualize into a pie chart the number of patients who had a stroke.


```python
df = pd.read_csv("healthcare-dataset-stroke-data.csv")
df.head(10)
```

```python
NumOfPatients=df["stroke"].size
StrokeNum=df["stroke"].sum()
print("Number of examples {}\nNumber of strokes {}".format(NumOfPatients,StrokeNum))

fig1, ax1 = plt.subplots()
ax1.pie([NumOfPatients ,StrokeNum], explode=(0,0.1), labels=["Total patients" , "Patients who had stroke"], autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  

plt.show()
```
